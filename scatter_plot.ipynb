```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
from sklearn.preprocessing import LabelEncoder

df = pd.read_excel(r"coin_toss.xlsx")       #reading from the dataset excel file
print(df)

initial=df.iloc[:,0]                        #1st column of the file is the initial dropping condition
x=df.iloc[:,1].values                       #2nd column of the file is the x-coordinates
y=df.iloc[:,2].values                       #3rd column of the file is the y-coordinates
distances=df.iloc[:,3]                      #4th column of the file is the distance from the origin
output=df.iloc[:,4]                         #5th column of the file is the output (heads or tails)

label_initial=LabelEncoder().fit_transform(initial)     #initial condition must be encoded as 0 and 1
label_output=LabelEncoder().fit_transform(output)       #output must be encoded as 0 and 1

plt.scatter(x, y, c=label_output)           #plotting the data points with colors corresponding to the output (yellow=tails, purple=heads)
plt.title("Coin positions")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

initial_tails_x=df.iloc[:35,1].values                   #first 35 rows is the condition when the coin is dropped with the tails up
initial_tails_y=df.iloc[:35,2].values
initial_tails_output=LabelEncoder().fit_transform(df.iloc[:35,4].values)

plt.scatter(initial_tails_x, initial_tails_y, c=initial_tails_output)           #plot of the positions of the coin with the first initial condition (dropped with the tails up)
plt.title("position where the coin falls when dropped with the tails up")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

initial_heads_x=df.iloc[35:70,1].values                 #second 35 rows is the condition when the coin is dropped with the heads up
initial_heads_y=df.iloc[35:70,2].values
initial_heads_output=LabelEncoder().fit_transform(df.iloc[35:70,4].values)

plt.scatter(initial_heads_x, initial_heads_y, c=initial_heads_output)           #plot of the positions of the coin with the second initial condition (dropped with the heads up)
plt.title("position where the coin falls when dropped with the heads up")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

initial_vertical_x=df.iloc[70:105,1].values             #last 35 rows is the condition when the coin is dropped vertically
initial_vertical_y=df.iloc[70:105,2].values
initial_vertical_output=LabelEncoder().fit_transform(df.iloc[70:105,4].values)

plt.scatter(initial_vertical_x, initial_vertical_y, c=initial_vertical_output)      #plot of the positions of the coin with the third initial condition (dropped vertically)
plt.title("position where the coin falls when dropped vertically")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()



```
