```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder

dataset = pd.read_excel(r"coin_toss.xlsx")
print(dataset)

x = dataset.iloc[:,1].values                #2nd column of the file is the x-coordinates
y = dataset.iloc[:,2].values                #3rd column of the file is the y-coordinates
distances = dataset.iloc[:,3].values        #4th column of the file is the distance from the origin
output=dataset.iloc[:,4].values
label_output=LabelEncoder().fit_transform(output)

sorted_distances = np.sort(distances)       #sorting the distances in ascending order

percentile_index = int(0.8 * len(sorted_distances)) #index of corresponding 80% of the distances

distances_80_percent = sorted_distances[:percentile_index] # the first percentile_index elements of the array

radius = np.max(distances_80_percent)           #the maximum value of the array corresponds the radius that contains 80% of the data points

circle = plt.Circle((0, 0), radius)             #plotting the circle with radius "radius" centered at the origin
fig, ax = plt.subplots()                        #creates a new figure
ax.add_patch(circle)                            #circle will be displayed on the plot

plt.scatter(x, y, c=label_output)
plt.title('The circle which contains 80% of the data points')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

print(f'The radius of the circle containing 80% of the points is: {radius}')
```
